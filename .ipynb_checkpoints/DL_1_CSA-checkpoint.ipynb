```python
import os
import numpy as np

x_train, x_test = np.array(list(range(101,1000+1))) , np.array(list(range(1,100+1)))

# train_size, test_size = 10000, 1000
# x_train, x_test = np.random.randint(101,1000+1,train_size), np.random.randint(1,100+1,test_size)

def do_bin(x,ln=10):
    st = bin(x)[2:]
    st = (ln-len(st))*'0' + st
    return [int(x) for x in st]

def do_sep(x,ln=4):
    st = str(x)
    st = (ln-len(st))*'0' + st
    return [int(x) for x in st]

x_train_bin, x_test_bin = np.array([do_bin(x) for x in x_train]), np.array([do_bin(x) for x in x_test])
x_train_sep, x_test_sep = np.array([do_sep(x) for x in x_train]), np.array([do_sep(x) for x in x_test])
x_train, x_test = x_train.reshape(len(x_train),-1), x_test.reshape(len(x_test),-1)

y_train, y_test = [], []
for x in x_train:
    x = x[0]
    if (not x%5) and (not x%3):
        y_train.append([1,0,0,0])
    elif (not x%3):
        y_train.append([0,1,0,0])
    elif (not x%5):
        y_train.append([0,0,1,0])
    else:
        y_train.append([0,0,0,1])

for x in x_test:
    x = x[0]
    if (not x%5) and (not x%3):
        y_test.append([1,0,0,0])
    elif (not x%3):
        y_test.append([0,1,0,0])
    elif (not x%5):
        y_test.append([0,0,1,0])
    else:
        y_test.append([0,0,0,1])
y_train, y_test = np.array(y_train), np.array(y_test)
```

```python
import tensorflow as tf
import tensorflow.keras as keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation

save_dir = os.path.join(os.getcwd(), 'saved_models')

epochs = 30
batch_size = 8
```

```python
def my_model(input_shape,output_shape):
    model = Sequential()
    model.add(Dense(512,input_dim=input_shape[1]))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(256))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add( Dense(128))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(output_shape[1]))
    model.add(Activation('softmax'))
    return model

model_name = 'integer_model.h5'
int_model = my_model(x_train.shape,y_train.shape)

# initiate RMSprop optimizer
# opt = keras.optimizers.RMSprop(learning_rate=0.0001, decay=1e-6)
opt = keras.optimizers.Adam()

# Let's train the model using RMSprop
int_model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])

int_model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(x_test, y_test),
          shuffle=True)

# Save model and weights
# if not os.path.isdir(save_dir):
#     os.makedirs(save_dir)
# model_path = os.path.join(save_dir, model_name)
# model.save(model_path)
# print('Saved trained model at %s ' % model_path)

# Score trained model.
scores = int_model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```

```python
def my_model(input_shape,output_shape):
    model = Sequential()
    model.add(Dense(128*4,input_dim=input_shape[1]))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(64*4))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add( Dense(32*4))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(output_shape[1]))
    model.add(Activation('softmax'))
    return model

model_name = 'sep_model.h5'
sep_model = my_model(x_train_sep.shape,y_train.shape)

# initiate RMSprop optimizer
# opt = keras.optimizers.RMSprop(learning_rate=0.0001, decay=1e-6)
opt = keras.optimizers.Adam(learning_rate=0.001, decay=0.0)

# Let's train the model using RMSprop
sep_model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])

sep_model.fit(x_train_sep, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(x_test_sep, y_test),
          shuffle=True)

# Save model and weights
# if not os.path.isdir(save_dir):
#     os.makedirs(save_dir)
# model_path = os.path.join(save_dir, model_name)
# model.save(model_path)
# print('Saved trained model at %s ' % model_path)

# Score trained model.
scores = sep_model.evaluate(x_test_sep, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```
